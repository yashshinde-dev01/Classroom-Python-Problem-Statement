```python
# '''Handling  Missing Data'''

# df.dropna(axis =0 , inplace=True)     -->axis =0 (remove rows)  -->axis=1 (remove  columns)

import pandas as pd
data={
    "Name":['Prasad',None,"Sanket","Prathamesh","Atharav","Rahul","Sojal","Pratik"],
    "Age":[10,None,13,15,12,15,14,14],
    "Salary":[100,None,103,105,103,104,106,104],
    "Performance Score":[85,None,45,26,49,12,79,99]
}

df=pd.DataFrame(data)
print(df)

df.dropna(axis=0,inplace=True)
print(df)
```

```python
'''fillna(value , inplace=True)'''
import pandas as pd
data={
    "Name":['Prasad',None,"Sanket","Prathamesh","Atharav","Rahul","Sojal","Pratik"],
    "Age":[10,None,13,15,12,15,14,14],
    "Salary":[100,None,103,105,103,104,106,104],
    "Performance Score":[85,None,45,26,49,12,79,99]
}

df=pd.DataFrame(data)
# print(df)

# df.fillna( 0 , inplace=True)

df["Age"].fillna(df["Age"].mean() , inplace=True)
df["Salary"].fillna(df["Salary"].mean() , inplace=True)
print(df)
```

```python
import pandas as pd
data={
    "Time":[1,2,3,4,5],
    "Value":[10,None,30,40,None]
}

df=pd.DataFrame(data)
print("Before interpretation")
print(df)

print("After interpretation")
df["Value"]=df["Value"].interpolate(method="linear")
print(df)

#After interpretation
#    Time  Value
# 0     1   10.0
# 1     2   20.0     <---
# 2     3   30.0
# 3     4   40.0
# 4     5   40.0    <--
```
