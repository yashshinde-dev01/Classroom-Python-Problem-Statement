```python
##### import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

FILE_NAME = "farm_records.csv"

# Create CSV file if not exists
if not os.path.exists(FILE_NAME):
    df = pd.DataFrame(columns=[
        "Farmer_Name", "Crop", "Soil_pH", "Rainfall",
        "Temperature", "Area", "Yield", "Water_Used", "Fertilizer"
    ])
    df.to_csv(FILE_NAME, index=False)


# ---------------- ADD RECORD ----------------
def add_record():
    name = input("Enter Farmer Name: ")
    crop = input("Enter Crop Name: ")
    ph = float(input("Enter Soil pH: "))
    rainfall = float(input("Enter Rainfall (mm): "))
    temp = float(input("Enter Temperature (°C): "))
    area = float(input("Enter Farm Area (acres): "))
    yield_amt = float(input("Enter Yield (kg): "))
    water = float(input("Enter Water Used (liters): "))
    fertilizer = input("Enter Fertilizer Used: ")

    new_data = pd.DataFrame({
        "Farmer_Name": [name],
        "Crop": [crop],
        "Soil_pH": [ph],
        "Rainfall": [rainfall],
        "Temperature": [temp],
        "Area": [area],
        "Yield": [yield_amt],
        "Water_Used": [water],
        "Fertilizer": [fertilizer]
    })

    df = pd.read_csv(FILE_NAME)
    df = pd.concat([df, new_data], ignore_index=True)
    df.to_csv(FILE_NAME, index=False)

    print("Record added successfully!")


# ---------------- VIEW RECORDS ----------------
def view_records():
    df = pd.read_csv(FILE_NAME)

    if df.empty:
        print("No records found.")
    else:
        print("\nFarm Records:")
        print(df)


# ---------------- DELETE RECORD ----------------
def delete_record():
    df = pd.read_csv(FILE_NAME)

    if df.empty:
        print("No records to delete.")
        return

    name = input("Enter Farmer Name to delete record: ")

    if name in df["Farmer_Name"].values:
        df = df[df["Farmer_Name"] != name]
        df.to_csv(FILE_NAME, index=False)
        print("Record deleted successfully!")
    else:
        print("Farmer not found.")


# ---------------- CROP RECOMMENDATION ----------------
def crop_recommendation():
    ph = float(input("Enter Soil pH: "))
    rainfall = float(input("Enter Rainfall (mm): "))
    temp = float(input("Enter Temperature (°C): "))

    if 6 <= ph <= 7 and rainfall > 700 and temp > 25:
        print("Recommended Crop: Rice")
    elif 6 <= ph <= 7.5 and rainfall < 500:
        print("Recommended Crop: Wheat")
    elif rainfall >= 500 and temp >= 20:
        print("Recommended Crop: Maize")
    else:
        print("Recommended Crop: Pulses")


# ---------------- WATER CALCULATOR ----------------
def water_calculator():
    crop = input("Enter Crop Name: ").lower()
    area = float(input("Enter Area (acres): "))

    water_data = {
        "rice": 5000,
        "wheat": 3000,
        "maize": 3500,
        "pulses": 2000
    }

    if crop in water_data:
        total = water_data[crop] * area
        print(f"Estimated Water Required: {total} liters")
    else:
        print("Crop data not available.")


# ---------------- FERTILIZER ADVISOR ----------------
def fertilizer_advisor():
    ph = float(input("Enter Soil pH: "))
    nitrogen = float(input("Enter Nitrogen level (0-100): "))

    if ph < 6:
        print("Suggestion: Add Lime to reduce acidity.")
    elif nitrogen < 40:
        print("Suggestion: Use Nitrogen-rich fertilizer (Urea).")
    else:
        print("Soil condition looks balanced.")


# ---------------- DATA ANALYSIS ----------------
def analyze_data():
    df = pd.read_csv(FILE_NAME)

    if df.empty:
        print("No data available.")
        return

    print("\n===== Analysis =====")
    print("Average Yield:", np.mean(df["Yield"]))
    print("Maximum Yield:", np.max(df["Yield"]))
    print("Average Rainfall:", np.mean(df["Rainfall"]))
    print("Average Temperature:", np.mean(df["Temperature"]))
    print("Average Water Used:", np.mean(df["Water_Used"]))


# ---------------- GRAPHS ----------------
def show_graphs():
    df = pd.read_csv(FILE_NAME)

    if df.empty:
        print("No data to plot.")
        return

    crop_yield = df.groupby("Crop")["Yield"].sum()

    plt.figure(figsize=(8, 5))
    plt.bar(crop_yield.index, crop_yield.values)
    plt.title("Crop Yield Analysis")
    plt.xlabel("Crop")
    plt.ylabel("Yield (kg)")
    plt.show()


# ---------------- MENU ----------------
while True:
    print("\n===== SMART AGRICULTURE MANAGEMENT SYSTEM =====")
    print("1. Add Farm Record")
    print("2. View Records")
    print("3. Delete Record")
    print("4. Crop Recommendation")
    print("5. Water Calculator")
    print("6. Fertilizer Advisor")
    print("7. Analyze Data")
    print("8. Show Graph")
    print("9. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_record()
    elif choice == "2":
        view_records()
    elif choice == "3":
        delete_record()
    elif choice == "4":
        crop_recommendation()
    elif choice == "5":
        water_calculator()
    elif choice == "6":
        fertilizer_advisor()
    elif choice == "7":
        analyze_data()
    elif choice == "8":
        show_graphs()
    elif choice == "9":
        print("Thank you for using the system!")
        break
    else:
        print("Invalid choice.")
```
